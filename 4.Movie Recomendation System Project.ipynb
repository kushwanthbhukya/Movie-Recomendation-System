```python
#creating a helper function
import ast
#ast? is used to convert the string to a list
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        #litral_eval? is used to convert the string to a list
        
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
#convert? is used to convert the string to a list
# why to use convert? because i need to convert the string to a list

movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj): #counter? is used to count the number of times a word occurs in a string
    l=[]
    counter=0
    for i in ast.literal_eval(obj): 
        #litral_eval? is used to convert the string to a list
        if counter!=3: 
            #what we get if counter!=3? 
            #if counter!=3 then we will append the name to the list
            l.append(i['name'])
            counter+=1  
        else:
            break       
    return l
```

```python

movies['cast']=movies['cast'].apply(convert3)
```

```python
#extracting the director name in the crew dataframe.
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
        
```

```python
from nltk.stem.porter import PorterStemmer
#stem? is used to convert the words to root words.
#porter stemmer? is used to convert the words to root words.
ps=PorterStemmer()

```

```python
def stem(text): 
    y=[ ]
    for i in text.split(): #textsplit? is used to split the text into words.
        y.append(ps.stem(i))

    return " ".join(y)
```
